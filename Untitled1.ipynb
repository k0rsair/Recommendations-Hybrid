# Этап 2: базовые и контентные модели

```python
#Неперсонализированная модель Popularity

class PopularityRecommender:
    
    def __init__(self,books, min_count_ratings=10):
        self.min_count_ratings = min_count_ratings
        self.popular_books = None
        self.books=books
        
    def fit(self, ratings):
        # Вычисляем средний рейтинг с минимальным количеством оценок
        book_stats = ratings.groupby('book_id').agg({
            'rating': ['count', 'mean']
        }).round(2)
        book_stats.columns = ['count', 'mean_rating']
        # Фильтруем по минимальному количеству оценок
        qualified_books = book_stats[book_stats['count'] >= self.min_count_ratings]
        # Сортируем по среднему рейтингу
        self.popular_books = qualified_books.sort_values('mean_rating', ascending=False)
        # Добавляем информацию о книгах
        self.popular_books = self.popular_books.merge(
            self.books[['book_id', 'original_title','title']], 
            on='book_id'
        )
        
    def recommend(self, n=10):
        return self.popular_books.head(n)

# Инициализация и обучение модели популярности
pop_model = PopularityRecommender(books=books,min_count_ratings=7000)
pop_model.fit(ratings)
print("Топ-10 популярных книг:")
display(pop_model.recommend(10))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

class ContentBasedRecommender:

    def __init__(self, book_tags, tags):
        self.tfidf_matrix = None
        self.vectorizer = None
        self.book_ids = None
        self.book_tags = book_tags
        self.tags = tags
        
    def prepare_book_profiles(self, books):
        # Создаем текстовые профили для книг
        book_profiles = []
        
        # Объединяем теги с названиями, группируем по id связи с книгой и объединяем имена тегов через пробел для создания профиля
        book_full_tags = self.book_tags.merge(self.tags, on='tag_id').groupby('goodreads_book_id').agg({'tag_name':' '.join})
        book_profiles = books.merge(book_full_tags,on='goodreads_book_id')
        book_profiles['original_title'] = book_profiles['original_title'].fillna(book_profiles['title'])
        book_profiles['profile'] = '' + book_profiles['original_title'] + ' ' + book_profiles['tag_name']
        return book_profiles[['profile','book_id']]
    
    def fit(self, books):
        # Подготавливаем профили
        book_profiles = self.prepare_book_profiles(books)

        # Векторизуем с помощью TF-IDF
        self.vectorizer = TfidfVectorizer(
            max_features=5000,
            stop_words='english',
            ngram_range=(1, 2)
        )
        
        self.tfidf_matrix = self.vectorizer.fit_transform(book_profiles['profile'])
        self.book_ids = book_profiles['book_id'].values
        
    def get_similar_books(self, book_id, n=5):
            
        # Находим индекс книги
        book_idx = np.where(self.book_ids == book_id)[0][0]

        # Вычисляем косинусную схожесть
        cosine_similarities = cosine_similarity(
            self.tfidf_matrix[book_idx:book_idx+1], 
            self.tfidf_matrix
        ).flatten()
        
        # Получаем индексы наиболее похожих книг (исключая саму книгу)
        similar_indexes = cosine_similarities.argsort()[::-1][1:n+1]

        similar_books = []
        
        # Индексируем книги для быстрого получения информации о книгах
        book_indexes = books.set_index('book_id')
        for idx in similar_indexes:
            similar_book_id = self.book_ids[idx]
            similarity_score = cosine_similarities[idx]
            
            # Получаем информацию о книге
            book_info = book_indexes.loc[similar_book_id]
            similar_books.append({
                'book_id': similar_book_id,
                'title': book_info['original_title'] if pd.notna(book_info['original_title']) else book_info['title'],
                'similarity': similarity_score
            })
        
        return pd.DataFrame(similar_books)

# Инициализация и обучение контентной модели
content_model = ContentBasedRecommender(book_tags, tags)
content_model.fit(books)

# Тестируем на примере
book = books.iloc[0]
print(f"Похожие книги на '{book['original_title']}':")
similar_books = content_model.get_similar_books(book['book_id'], 10)
print(similar_books[['title', 'similarity']])
```

# Этап 3: коллаборативная фильтрация (Collaborative Filtering — CF)

```python
import random

class ItemBasedCF:
    def __init__(self, k=5):
        self.k = k
        self.user_item_matrix = None
        self.similarity_matrix = None
        self.books_rated = None
         
    def fit(self, ratings):
        # Матрица пользователь-книга
        self.user_item_matrix = ratings.pivot_table(
            index='user_id', 
            columns='book_id', 
            values='rating'
        ).fillna(0)

        # Матрица схожести книг
        self.similarity_matrix = cosine_similarity(self.user_item_matrix.T)
        self.similarity_df = pd.DataFrame(
            self.similarity_matrix,
            index=self.user_item_matrix.columns,
            columns=self.user_item_matrix.columns
        )
        # Какие книги оценивал пользователь
        self.books_rated = ratings.groupby('user_id')['book_id'].apply(set)

    def predict_rating(self, user_id, book_id):
        if user_id not in self.user_item_matrix.index or book_id not in self.similarity_df.columns:
            return 0

        user_ratings = self.user_item_matrix.loc[user_id]
        book_similarities = self.similarity_df[book_id]
        rated_books = self.books_rated[user_id]

        valid_books = [b for b in rated_books if b != book_id and user_ratings[b] > 0]
        if not valid_books:
            return 0
        
        similar_ratings = [(book_similarities[b], user_ratings[b]) for b in valid_books]
        similar_ratings = sorted(similar_ratings, reverse=True)[:self.k]

        if not similar_ratings:
            return 0

        similarities, ratings = zip(*similar_ratings)
        weighted_sum = np.dot(similarities, ratings)
        sum_weights = np.sum(np.abs(similarities))
        return round(weighted_sum / sum_weights if sum_weights > 0 else 0, 2)

    def recommend(self, user_id, all_books, k=10, candidate_size=1000):
        """Рекомендует top-k книг для пользователя"""
        if user_id not in self.user_item_matrix.index:
            return []
        
        rated = self.books_rated[user_id]
        candidates = list(set(all_books) - rated)

        # Случайный поднабор кандидатов (чтобы ускорить и не сравнивать со всеми)
        if len(candidates) > candidate_size:
            candidates = random.sample(candidates, candidate_size)

        preds = {b: self.predict_rating(user_id, b) for b in candidates}
        return sorted(preds, key=preds.get, reverse=True)[:k]
```

### В чём вычислительная сложность этого подхода? Как его можно оптимизировать для больших данных?
У этого подхода большая проблема в том, что его сложность растет пропорционально квадрату количества книг помноженное на количество пользователей, так как мы строим матрицу сравнения каждой книги с каждой по оценкам каждого пользователя. С каждым новым пользователем и каждой добавленной новой книгой строить матрицы становится все труднее. 
Оптимизация для больших данных: 
1. можно кешировать матрицы схожести, раз в какое-то время, чтобы вычисление их не происходило каждый раз.
2. можно ограничивать выборку книг, беря например книги того же жанра, года, автора.

### Этап 4: матричные разложения (Matrix Factorization)

```python
from surprise import Dataset, Reader, SVD, accuracy
from surprise.model_selection import train_test_split

class MatrixFactorizationRecommender:
    
    def __init__(self, books, n_factors=50, n_epochs=20, lr=0.005, reg=0.02):
        self.n_factors = n_factors
        self.n_epochs = n_epochs
        self.lr = lr
        self.reg = reg
        # Инициализируем модель SVD
        self.model = SVD(
            n_factors=self.n_factors,
            n_epochs=self.n_epochs,
            lr_all=self.lr,
            reg_all=self.reg
        )
        self.ratings = None
        self.books = books

    
    def fit(self, train_set):
        data = Dataset.load_from_df(train_set[['user_id', 'book_id', 'rating']], reader)
        self.model.fit(data.build_full_trainset())
        self.ratings = train_set
        
    def predict_rating(self, user_id, book_id):
        try:
            prediction = self.model.predict(user_id, book_id)
            return prediction.est
        except:
            return 0
    
    def get_recommendations(self, user_id, n=5):
        # Получаем все книги, которые пользователь не оценивал
        rated_books = set(self.ratings[self.ratings['user_id'] == user_id]['book_id'])
        unrated_books = set(self.books['book_id']) - rated_books
        
        # Предсказываем рейтинги
        predictions = []
        for book_id in list(unrated_books):
            pred_rating = self.predict_rating(user_id, book_id)
            predictions.append((book_id, pred_rating))
        
        # Сортируем по предсказанному рейтингу
        predictions.sort(key=lambda x: x[1], reverse=True)
        
        # Добавляем информацию о книгах
        recommendations = []
        book_indexed = self.books.set_index('book_id')
        for book_id, rating in predictions[:n]:
            book_info = book_indexed.loc[book_id]
            recommendations.append({
                'book_id': book_id,
                'title': book_info['original_title'],
                'predicted_rating': rating
            })
        
        return pd.DataFrame(recommendations)

 # Подготавливаем данные для Surprise
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(ratings[['user_id', 'book_id', 'rating']], reader)

train_set, test_set = train_test_split(data, test_size=0.2, random_state=11)

mf_model = MatrixFactorizationRecommender(books, n_factors=20, n_epochs=10)
mf_model.fit(train_set)

predictions = mf_model.model.test(test_set)
rmse = accuracy.rmse(predictions, verbose=False)

print(f"Оценка RMSE {rmse}")

test_user = ratings.loc[0]['user_id']
print(f"Рекомендации SVD для пользователя {test_user}:")
svd_recommendations = mf_model.get_recommendations(test_user, 5)
print(svd_recommendations[['title', 'predicted_rating']])
```

```python
# Разделяем данные на train и test

def split_right_train_test(rating_df,threshhold = 4, k = 5):
    # Выделим только вадные оценки
    v_rating = rating_df[rating_df['rating'] >= threshhold]
    # Отделяем пользователей у которых есть валидные оценки больше или равное нужному количеству
    res = v_rating.groupby('user_id').agg({'rating':'count'}).reset_index()

    valid_users = set(res[res['rating'] >= k * 2]['user_id'])

    valid_data = rating_df[rating_df['user_id'].isin(valid_users)]
    
     # Фильтруем записи с оценкой >= 4
    high_ratings = valid_data[valid_data['rating'] >= 4].copy()
    
    # Группируем по пользователям и для каждого выбираем k случайных записей с оценкой >= 4
    test_high_ratings = (
        high_ratings
        .groupby('user_id')
        .apply(lambda x: x.sample(n=k, random_state=42))
        .reset_index(drop=True, level=[0])# Убираем мульти индекс сохарняя оригинальные индексы записей
    )
    test_indixes = test_high_ratings.index
    
    return valid_data.drop(test_indixes).reset_index(drop=True), test_high_ratings.reset_index(drop=True)
    
    
    
    
```

```python
def precision_at_k(recommended, relevant):
    if not recommended:
        return 0
    return len(set(recommended) & set(relevant)) / len(recommended)

def recall_at_k(recommended, relevant):
    if not relevant:
        return 0
    return len(set(recommended) & set(relevant)) / len(relevant)

def ndcg_at_k(recommended, relevant):
    import numpy as np
    dcg = 0.0
    for i, rec in enumerate(recommended):
        if rec in relevant:
            dcg += 1 / np.log2(i + 2)
    idcg = sum(1 / np.log2(i + 2) for i in range(min(len(relevant), len(recommended))))
    return dcg / idcg if idcg > 0 else 0


model = ItemBasedCF(k=5)
model.fit(train)
all_books = ratings['book_id'].unique()

# Словарь: user_id -> релевантные книги (оценка >= 4 в тесте)
relevant_books = (
    test[test['rating'] >= 4]
    .groupby('user_id')['book_id']
    .apply(set)
    .to_dict()
)

precisions, recalls, ndcgs = [], [], []

for user_id, relevant in relevant_books.items():
    recommended = model.recommend(user_id, all_books, k=10, candidate_size=1000)
    if not recommended:
        continue
    precisions.append(precision_at_k(recommended, relevant))
    recalls.append(recall_at_k(recommended, relevant))
    ndcgs.append(ndcg_at_k(recommended, relevant))

print('Precision@10:', np.mean(precisions))
print('Recall@10:', np.mean(recalls))
print('NDCG@10:', np.mean(ndcgs))
```

```python
def precision_recall_at_k(model, test, all_books, k=10, threshold=3.5):
    precisions = []
    recalls = []

    for user_id in test['user_id'].unique():
        # Истинно релевантные книги (в test)
        user_test = test[test['user_id'] == user_id]
        true_relevant = set(user_test[user_test['rating'] >= threshold]['book_id'])
        if not true_relevant:
            continue

        # Рекомендации модели
        recs = model.recommend(user_id, all_books, n=k)
        recommended_books = {book_id for book_id, _ in recs}

        # Пересечение
        hits = len(true_relevant & recommended_books)

        precisions.append(hits / k)
        recalls.append(hits / len(true_relevant))

    return np.mean(precisions), np.mean(recalls)
```

```python
from scipy.sparse import csr_matrix

class CollaborativeFilteringRecommender:
    """Item-Based Collaborative Filtering с явными оценками и косинусной похожестью."""

    def __init__(self, k_neighbors=30, marks_treshold=100, min_ratings_for_similarity=5):
        self.k_neighbors = k_neighbors
        self.marks_treshold = marks_treshold
        self.min_ratings_for_similarity = min_ratings_for_similarity
        self.popular_items_stats = None
        self.interaction_matrix_explicit_sparse = None
        self.item_similarity_explicit = None
        self.user_index_map = None
        self.book_index_map = None
        self.index_user_map = None # Обратный маппинг
        self.index_book_map = None # Обратный маппинг
        self.global_mean_rating = None # Средний рейтинг по всем данным

    def fit(self, train_data):
        """
        Обучает модель Item-Based CF.

        Строит разреженную матрицу взаимодействий, рассчитывает схожесть между
        книгами и вычисляет статистику популярности для fallback.

        Args:
            train_data (pd.DataFrame): DataFrame, содержащий как минимум
                колонки 'user_id', 'book_id', 'rating'.
        """
        print(f"обучение item-item cf (k_neighbors={self.k_neighbors})...")

        # Для использования scipy.sparse нам понадобятся user_id и book_id в виде последовательных индексов
        train_data_sparse = train_data.copy()
        train_data_sparse['user_id_indexed'] = train_data_sparse['user_id'].astype('category').cat.codes
        train_data_sparse['book_id_indexed'] = train_data_sparse['book_id'].astype('category').cat.codes

        # Создаем словари сопоставления для преобразования исходных ID в индексы и обратно
        self.user_index_map = train_data_sparse.set_index('user_id')['user_id_indexed'].drop_duplicates()
        self.book_index_map = train_data_sparse.set_index('book_id')['book_id_indexed'].drop_duplicates()
        self.index_user_map = pd.Series(self.user_index_map.index, index=self.user_index_map.values)
        self.index_book_map = pd.Series(self.book_index_map.index, index=self.book_index_map.values)


        # Построение разреженной матрицы взаимодействий с явными оценками
        self.interaction_matrix_explicit_sparse = csr_matrix(
            (train_data_sparse['rating'], (train_data_sparse['user_id_indexed'], train_data_sparse['book_id_indexed']))
        )

        # Получаем среднюю оценку по всем данным для использования в случае отсутствия оценок
        self.global_mean_rating = train_data_sparse['rating'].mean()


        # Рассчитываем матрицу попарной схожести между книгами с использованием косинусной близости
        # Транспонируем матрицу взаимодействий, чтобы книги стали строками, а пользователи - столбцами
        item_user_matrix = self.interaction_matrix_explicit_sparse.T

        # Рассчитываем косинусную схожесть между книгами
        # Поскольку item_user_matrix разреженная, cosine_similarity будет работать эффективно
        self.item_similarity_explicit = cosine_similarity(item_user_matrix)


        # Вычисляем статистику популярности и применяем порог (для fallback)
        self.popular_items_stats = train_data.groupby('book_id')['rating'].agg(['mean', 'count']).reset_index()
        self.popular_items_stats = self.popular_items_stats[self.popular_items_stats['count'] >= self.marks_treshold].copy()

        # Добавляем столбец с book_id, отсортированными по популярности, для использования в fallback рекомендациях
        self.popular_items_stats['book_id_sorted'] = self.popular_items_stats['book_id'].tolist()



    def predict_rating(self, user_id, book_id):
        """
        Предсказывает оценку пользователя для заданной книги на основе Item-Based CF.
        Используется для ранжирования кандидатов в методе recommend.
        """
        try:
            user_idx = self.user_index_map[user_id]
            book_idx = self.book_index_map[book_id]
        except KeyError:
            # Если пользователя или книги нет в обучающих данных, возвращаем средний рейтинг
            return self.global_mean_rating

        book_similarity_vector = self.item_similarity_explicit[book_idx]

        # Исключаем саму книгу из списка похожих и получаем индексы K наиболее похожих
        book_similarity_vector_copy = book_similarity_vector.copy()
        book_similarity_vector_copy[book_idx] = -1.0 # Исключаем саму книгу
        # Получаем индексы K наиболее похожих книг из всего каталога (включая те, что пользователь мог видеть)
        similar_books_indices = np.argsort(book_similarity_vector_copy)[::-1][:self.k_neighbors]

        # Получаем оценки пользователя для этих K похожих книг из матрицы взаимодействий
        # Используем .indices и .data для работы с разреженной строкой
        user_ratings = self.interaction_matrix_explicit_sparse[user_idx]
        rated_indices = user_ratings.indices # Индексы книг, которые пользователь оценил
        rated_values = user_ratings.data    # Оценки пользователя для этих книг


        # Находим пересечение похожих книг и книг, оцененных пользователем
        common_rated_indices_mask = np.isin(similar_books_indices, rated_indices)
        common_similar_indices = similar_books_indices[common_rated_indices_mask]

        if common_similar_indices.size == 0:
            # Если пользователь не оценил ни одну из K похожих книг, возвращаем средний рейтинг
            return self.global_mean_rating
        else:
            # Получаем индексы оценок пользователя для общих книг
            rated_indices_map = {idx: i for i, idx in enumerate(rated_indices)}
            user_ratings_for_common_books = np.array([rated_values[rated_indices_map[idx]] for idx in common_similar_indices])


            # Получаем значения схожести для этих общих книг
            similarity_values_for_common_books = book_similarity_vector[common_similar_indices]

            # Применяем фильтр по минимальной схожести, если необходимо (можно добавить параметр)
            # Например: valid_indices = similarity_values_for_common_books > 0.0


            # Вычисляем взвешенную сумму оценок
            weighted_sum = np.sum(user_ratings_for_common_books * similarity_values_for_common_books)
            sum_of_weights = np.sum(np.abs(similarity_values_for_common_books)) # Использование абсолютного значения для весов

            if sum_of_weights == 0:
                 # Если сумма весов равна 0 (например, если все схожести отрицательные или 0),
                 # возвращаем средний рейтинг
                return self.global_mean_rating
            else:
                predicted_rating = weighted_sum / sum_of_weights
                return predicted_rating


    def recommend(self, user_id, k=10, user_seen_items_dict=None, all_items=None):
        if self.interaction_matrix_explicit_sparse is None or self.item_similarity_explicit is None:
            print("Модель Item-Based CF не обучена. Пожалуйста, вызовите метод fit сначала.")
            # Возвращаем популярные книги в качестве запасного варианта
            return self._fallback_recommendations(set(), k)

        # Получаем набор книг, которые пользователь уже видел/оценил
        # Используем user_seen_items_dict, если он передан (для оценки метрик)
        # Иначе получаем из train_data_sparse
        if user_seen_items_dict is not None:
            user_seen_items = user_seen_items_dict.get(user_id, set())
        else:
             # Получаем оценки пользователя из обучающих данных
            try:
                user_idx = self.user_index_map[user_id]
                user_ratings = self.interaction_matrix_explicit_sparse[user_idx]
                user_seen_items = {self.index_book_map[idx] for idx in user_ratings.indices}
            except KeyError:
                # Если пользователя нет в обучающих данных (холодный старт),
                # возвращаем популярные книги
                print(f"Пользователь {user_id} не найден в обучающих данных. Возвращаем популярные книги.")
                return self._fallback_recommendations(set(), k)


        seen_set = user_seen_items # Используем набор для быстрого поиска
        # user_seen_items_list = list(user_seen_items) # Не используется в этой версии

        all_item_ids = all_items if all_items is not None else self.book_index_map.index.tolist()


        # Находим книги, которые пользователь еще не видел и которые прошли порог популярности
        # Убрана проверка на books_profiles, так как она не относится к CF по оценкам
        # ОПТИМИЗАЦИЯ: Вместо итерации по всем all_item_ids, будем агрегировать рекомендации
        # на основе просмотренных пользователем книг.

        display(self.interaction_matrix_explicit_sparse)
        user_rated_books_indices = self.interaction_matrix_explicit_sparse[self.user_index_map[user_id]].indices
        user_rated_books_ids = {self.index_book_map[idx] for idx in user_rated_books_indices}

        # Словарь для хранения суммарной взвешенной схожести для каждой потенциальной рекомендуемой книги
        # Ключ: book_id кандидата, Значение: суммарная взвешенная схожесть
        candidate_scores = {}

        for rated_book_id in user_rated_books_ids:
            try:
                rated_book_idx = self.book_index_map[rated_book_id]
            except KeyError:
                # Пропускаем книги, которых нет в маппинге (хотя такого быть не должно после fit)
                continue

            # Получаем вектор схожести для оцененной пользователем книги
            similarity_vector = self.item_similarity_explicit[rated_book_idx]

            # Находим k_neighbors наиболее похожих книг (включая саму оцененную книгу)
            # Исключаем саму книгу и уже оцененные пользователем книги
            similar_books_indices = np.argsort(similarity_vector)[::-1][:self.k_neighbors + 1] # +1 чтобы исключить саму книгу

            for i in similar_books_indices:
                similar_book_id = self.index_book_map[i] # Получаем book_id по индексу
                similarity_score = similarity_vector[i]

                # Исключаем саму оцененную книгу, книги, которые пользователь уже видел,
                # и книги, которые не прошли порог популярности (для кандидатов)
                if similar_book_id != rated_book_id and similar_book_id not in seen_set \
                   and similar_book_id in self.popular_items_stats['book_id'].values:

                    # Получаем оценку пользователя для rated_book_id
                    # Находим индекс rated_book_id в строке пользователя
                    user_ratings_indices = self.interaction_matrix_explicit_sparse[self.user_index_map[user_id]].indices
                    user_ratings_values = self.interaction_matrix_explicit_sparse[self.user_index_map[user_id]].data
                    rated_book_idx_in_user_ratings = np.where(user_ratings_indices == rated_book_idx)[0][0]
                    user_rating_for_rated_book = user_ratings_values[rated_book_idx_in_user_ratings]


                    # Взвешиваем схожесть на оценку пользователя (можно использовать разные схемы взвешивания)
                    # Например, просто суммировать схожесть, или умножать на оценку
                    weighted_similarity = similarity_score # Простой вариант: суммируем схожесть
                    # Или: weighted_similarity = similarity_score * user_rating_for_rated_book # Взвешивание по оценке

                    if similar_book_id not in candidate_scores:
                        candidate_scores[similar_book_id] = weighted_similarity
                    else:
                        candidate_scores[similar_book_id] += weighted_similarity # Суммируем взвешенную схожесть


        if candidate_scores:
            # Сортируем кандидатов по убыванию суммарной взвешенной схожести
            sorted_candidates = sorted(candidate_scores.items(), key=lambda item: item[1], reverse=True)
            # Берем топ-k книг
            recommendations = [item_id for item_id, score in sorted_candidates[:k]]
        else:
            recommendations = []

        # Если рекомендаций меньше k, дополняем их популярными книгами,
        # которые прошли порог и пользователь не видел
        if len(recommendations) < k:
             additional_recs = self._fallback_recommendations(
                 seen_set.union(set(recommendations)), k - len(recommendations))
             recommendations.extend(additional_recs)


        return recommendations


    def _fallback_recommendations(self, user_seen_items, k):
        """Возвращает популярные книги, прошедшие порог, в качестве запасного варианта."""
        if self.popular_items_stats is None or self.popular_items_stats.empty:
             return []

        seen_set = set(user_seen_items)
        recs = []
        # Итерируемся по book_id в отсортированном порядке популярности из popular_items_stats
        for book_id in self.popular_items_stats['book_id_sorted']:
             if book_id not in seen_set:
                 recs.append(book_id)
                 if len(recs) >= k:
                     break
        return recs
```
